```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
rng = np.random.default_rng()
```

```python
X = np.load("X.npy")
y = np.load("y.npy")

X_2 = np.load("X_2.npy")
y_2 = np.load("y_2.npy")

```

```python
channel = 7
y_0_index = np.where(y == 0)[0]
y_0 = y[y_0_index]
x_0 = X[y_0_index, channel]

y_1_index = np.where(y == 1)[0]
y_1 = y[y_1_index]
x_1 = X[y_1_index, channel]

x = X_2[:, channel]
y = y_2
```

```python
x_1.shape
```

```python
cmap = ['red', 'green', 'blue', 'yellow']
for i in range(0,50):
    fs = 250
    freq =125.0
    x_next = x[i]
    y_next = y[i]
    f, Pxx_den = signal.welch(x_next, fs, nperseg=50, noverlap=20, nfft = 1000)
    plt.semilogy(f, Pxx_den, label = str(y_next), c = cmap[y_next])
    plt.ylim([0.5e-3, 1])
    plt.xlim([0,60])
    plt.xlabel('frequency [Hz]')
    plt.ylabel('PSD [V**2/Hz]')
    # plt.show()
plt.legend()
plt.show()
```
